```

import imageio
frames = []
image_count = 0
while True:
    frame = img1[image_count]
    cv2.imshow("frame", frame)
    key = cv2.waitKey(0)
    if key == ord("a"):
        image_count += 1
        frames.append(frame)
        print( image_count)
    elif key == ord("q"):
        break
print("Adding new image:", len(frames))
```

```
from PIL import Image
import imageio
frames = []
for frame_number in range(1, len(fnames)):
  frame = Image.open(fnames[frame_number])
  frames.append(frame)
frames[0].save(
    'da.gif',
    save_all=True,
    append_images=frames[1:], 
    optimize=True,
    duration=100,
    loop=0
)
```
